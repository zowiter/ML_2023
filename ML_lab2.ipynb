```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("seattle-weather.csv")
df
```

```python
#Проверим, все ли данные корректно заполнены
df.isnull().sum()
```

```python
#Раздеим датасет на признаки (Х) и ответы (у)
#предсказывать будем прогнощ погоды (weather)
y = df['weather']
X = df.drop(['date', 'weather'], axis=1)
```

```python
# Теперь можно поделить на тестовую и валидационную выборку в соотношении 0.28
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.28)

print("data:   ", df.shape[0])
print("X_train:", X_train.shape[0])
print("X_test: ", X_test.shape[0])
```

```python
#k-ближайших соседей
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
#Наивная байесовская классификация
from sklearn.naive_bayes import GaussianNB

gauss = GaussianNB()
gauss.fit(X_train, y_train)
y_pred = gauss.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
#Стохастический градиентный спуск
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(n_jobs=-1) # Run using all processors
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_jobs=-1)
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)

accuracy_score(y_test, y_pred)
```
